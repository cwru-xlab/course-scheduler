## Question 1: Calculate Sum of Even Numbers

Write a function called `sum_even_numbers` that takes a list of integers as input and returns the sum of all even numbers in the list.

**Example:**
```python
sum_even_numbers([1, 2, 3, 4, 5, 6]) # Should return 12 (2 + 4 + 6)
sum_even_numbers([1, 3, 5]) # Should return 0
```

````python
# Student's answer with syntax error
def sum_even_numbers(numbers):
    total = 0
    for num in numbers
        if num % 2 == 0:
            total += num
    return total
````

````python
# Test cases for Question 1
assert sum_even_numbers([1, 2, 3, 4, 5, 6]) == 12, "Test case 1 failed"
assert sum_even_numbers([1, 3, 5]) == 0, "Test case 2 failed"
assert sum_even_numbers([2, 4, 6, 8]) == 20, "Test case 3 failed"
assert sum_even_numbers([]) == 0, "Test case 4 failed"
print("All test cases passed!")
````

## Question 2: Find Maximum Value

Write a function called `find_maximum` that takes a list of numbers and returns the maximum value. You may not use the built-in `max()` function.

**Example:**
```python
find_maximum([3, 7, 2, 9, 1]) # Should return 9
find_maximum([5]) # Should return 5
```

````python
# Student's answer with runtime error
def find_maximum(numbers):
    max_value = numbers[0]
    for num in numbers:
        if num > max_value:
            max_value = num
    return max_value
````

````python
# Test cases for Question 2
assert find_maximum([3, 7, 2, 9, 1]) == 9, "Test case 1 failed"
assert find_maximum([5]) == 5, "Test case 2 failed"
assert find_maximum([-1, -5, -3]) == -1, "Test case 3 failed"
assert find_maximum([]) == None, "Test case 4 failed - should handle empty list"
print("All test cases passed!")
````

## Question 3: Count Vowels

Write a function called `count_vowels` that takes a string as input and returns the number of vowels (a, e, i, o, u) in the string. The function should be case-insensitive.

**Example:**
```python
count_vowels("Hello World") # Should return 3
count_vowels("Python") # Should return 1
```

````python
# Student's correct answer
def count_vowels(text):
    vowels = 'aeiouAEIOU'
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count
````

````python
# Test cases for Question 3
assert count_vowels("Hello World") == 3, "Test case 1 failed"
assert count_vowels("Python") == 1, "Test case 2 failed"
assert count_vowels("AEIOU") == 5, "Test case 3 failed"
assert count_vowels("xyz") == 0, "Test case 4 failed"
assert count_vowels("") == 0, "Test case 5 failed"
print("All test cases passed!")
````
